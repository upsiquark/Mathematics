```python
#my own working solution to a function that yields
#the greatest common divisor(gcd) between two numbers
def gcd_calc(a,b):
    while b != 0:
        return gcd_calc(b,a%b)
    return a
```

```python
#more elegant solution that yields the gcd between two numbers
#and the bezout coefficients
def gcd(a,b):
    if a == 0:
        return b, 0, 1
    g, x0, y0 = gcd(b%a, a)
    x = y0 - (b//a) * x0
    y = x0
    return g, x, y
```
