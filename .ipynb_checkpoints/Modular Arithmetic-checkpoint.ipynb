```python
#Some functions for modular arithmetic without using the modulo operator!:D

#a function that returns remainders mod n
def mod_n_remainders(a,n):
    if isinstance(a/n, int):
        a = 0
    if (a>n):
        p = a//n
        a = a - p*n
    return a

#a function that returns the sum of two integers mod n
def mod_n_sum(a,b,n):
    a = mod_n_remainders(a,n)
    b = mod_n_remainders(b,n)
    s = a + b
    return mod_n_remainders(s,n)

#a function that returns the product of two integers mod n
def mod_n_product(a,b,n):
    a = mod_n_remainders(a,n)
    b = mod_n_remainders(b,n)
    m = a * b
    return mod_n_remainders(m,n)

#a function that returns the sum and product of two integers mod n
def mod_n_operations(a,b,n):
    a = mod_n_remainders(a,n)
    b = mod_n_remainders(b,n)
    m = a * b
    s = a + b

    return mod_n_remainders(m,n), mod_n_remainders(s,n)

#trustee gcd calc
def gcd(a,b):
    while b != 0:
        return gcd(b, mod_n_remainders(a,b))
    return a

#a function that returns the inverse of an integer mod n
#given the condition that the gcd between the integer and n is 1
def mod_n_identity(a,n):
    if gcd(a,n) == 1:
        for i in range(n):
            if mod_n_product(a,i,n) == 1:
                return i
```
