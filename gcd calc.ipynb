```python
#more elegant solution that yields the bezout coefficients
def gcd(a,b):
    if a == 0:
        return b, 0, 1
    g, x0, y0 = gcd(b%a, a)
    x = y0 - (b//a) * x0
    y = x0
    return g, x, y
```
